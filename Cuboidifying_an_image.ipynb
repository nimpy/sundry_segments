```python
import imageio
import matplotlib.pyplot as plt
import numpy as np
```

```python
# image = imageio.imread("/home/niaki/Pictures/vianden_castle.jpg")

image = imageio.imread("/home/niaki/Pictures/vianden_castle_intermediate_representation_gray1.jpg")
image = np.repeat(image, 3, axis=1).reshape((image.shape[0], image.shape[1], 3))  # if grayscale
```

```python
image_height = image.shape[0]
image_width = image.shape[1]
nr_channels = image.shape[2]
```

```python
size_diff = 70 # int(image_height * 0.2)

image3d_height = image_height + size_diff
image3d_width = image_width + size_diff

image3d = 255* np.ones((image3d_height, image3d_width, nr_channels), dtype=np.uint8)
```

```python
image3d[image3d_height - image_height : , 0 : image_width, :] = image
```

```python
plt.imshow(image3d)
```

```python
for w in range(image_width):
    for h in range(size_diff):
        for i in range(size_diff):
            image3d[size_diff - h, w + h, :] = image[0, w, :]
```

```python
for h in range(image_height):
    for w in range(size_diff):
        for i in range(size_diff):
            image3d[h + w, image_width - 1 + size_diff - w, :] = image[h, image_width - 1, :]
```

```python
imageio.imwrite("/home/niaki/Pictures/vianden_castle_intermediate_representation_gray1_3d.jpg", image3d)
```
