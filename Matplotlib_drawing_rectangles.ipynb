```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import imageio
```

```python
image = imageio.imread("Lenna.bmp")
```

```python
fig, ax = plt.subplots(1)
ax.imshow(image)
# col, row, rect_width, rect_height (OBRNUTO od ostalih biblioteka!)
rect = patches.Rectangle((10, 20), 100, 200, linewidth=1, edgecolor='r', facecolor='none')
ax.add_patch(rect)
plt.show()
```
